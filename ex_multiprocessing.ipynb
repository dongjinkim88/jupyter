```python
import concurrent.futures as cf
import time
import os

start = time.perf_counter()

def f(sec):
    print(f'process {os.getpid()}: sleeping {sec} seconds')
    time.sleep(sec)
    return f'done sleeping {sec} seconds'

num_process = 20
with cf.ProcessPoolExecutor() as executor:
#with cf.ThreadPoolExecutor() as executor:
#    secs = [5,4,3,2,1]
    secs = [2 for i in range(num_process)]
    results = executor.map(f, secs)
    
    for result in results:
        print(result)    

finish = time.perf_counter()
print(f'Finished in {round(finish-start,2)} seconds')
#Multiprocessing: Finished in 10.11 seconds
#Multithreading: Finished in 2.01 seconds (for 10 process)
#Multithreading: failed (for 20 process)
```

```python
import multiprocessing as mp
import time
import os

start = time.perf_counter()

def f(sec):
    print(f'process {os.getpid()}: sleeping {sec} seconds')
    time.sleep(sec)
    print(f'done sleeping {sec} seconds')

processes = []
num_process = 20
for _ in range(num_process):
    p = mp.Process(target=f, args=[2])
    p.start()
    processes.append(p)

for process in processes:
    process.join()
    
finish = time.perf_counter()
print(f'Finished in {round(finish-start,2)} seconds')
#Finished in 2.38 seconds
```
